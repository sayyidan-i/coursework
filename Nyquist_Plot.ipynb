```python
import control
import matplotlib.pyplot as plt
```

```python
# G(s) = 1/s  --> G(jw) = 1/jw

# as^2+bs+c --> (a, b, c)

G = control.TransferFunction((1), (1, 0))
w = [0.1, 100]
count = control.nyquist_plot(G, w)
```

```python
# G(jw)=1/(1+jw) --> (1), (1, 1)

G = control.TransferFunction((1), (1, 1, 0))
# sys = control.feedback(G,1);
sys = G
count = control.nyquist_plot(sys)
print(sys)
```

```python
import matplotlib.pyplot as plt

G = control.TransferFunction((1), (1, 1, 0))
w = [0.1, 10000000]
count = control.nyquist_plot(G, w)
plt.axis('equal')
plt.show()
```
